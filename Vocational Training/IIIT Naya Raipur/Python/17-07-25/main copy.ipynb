```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
from gammatone.gtgram import gtgram

# Load audio file
audio_url = "S-Voice Sample.wav"
y, sr = librosa.load(audio_url, sr=None)

print(f"Audio shape: {y.shape}")
print(f"Sample rate: {sr}")

# Plot waveform
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform from URL")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

# Compute GFCC-like features
window_time = 0.025  # 25ms window
hop_time = 0.010     # 10ms hop
fmin = 50            # Minimum frequency
channels = 64        # Number of frequency channels (you can adjust this)

# Compute Gammatone spectrogram
gfcc_like = gtgram(y, sr, window_time, hop_time, channels, fmin)
print("GFCC-like feature shape:", gfcc_like.shape)
```

```python
spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)
print("Spectral Centroid shape:", spectral_centroids.shape)
```

```python
f0 = librosa.yin(y, fmin=50, fmax=400, sr=sr)
avg_pitch_yin = np.mean(f0)
print(f"Average Pitch (YIN): {avg_pitch_yin:.2f} Hz")
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os

audio_file_path = "S-Voice Sample.wav"

try:
    y, sr = librosa.load(audio_file_path, sr=None)
    print(f"Audio loaded successfully! Sampling Rate (sr): {sr} Hz")
    print(f"Audio shape: {y.shape}")
    print(f"Audio duration: {len(y)/sr:.2f} seconds")

except FileNotFoundError:
    print(f"Error: Audio file not found at '{audio_file_path}'.")
    print("Please make sure the file is in the correct directory or provide the full path.")
except Exception as e:
    print(f"An error occurred while loading the audio file: {e}")

plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr, color='blue')
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

Y = np.fft.fft(y)
freqs = np.fft.fftfreq(len(Y), 1/sr)
magnitude = np.abs(Y)

plt.figure(figsize=(12, 6))
plt.plot(freqs[:len(freqs)//2], magnitude[:len(magnitude)//2])
plt.title("Frequency Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.grid()
plt.tight_layout()
plt.show()

n_mfcc_coeffs = 13
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc_coeffs)
print(f"MFCCs shape: {mfccs.shape}")

plt.figure(figsize=(12, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time', cmap='viridis')
plt.colorbar(label='MFCCs')
plt.title('MFCCs Over Time')
plt.xlabel('Time (s)')
plt.ylabel('MFCC Coefficient')
plt.tight_layout()
plt.show()

gfccs_placeholder_order = n_mfcc_coeffs
gfccs_mean_placeholder = np.zeros(gfccs_placeholder_order)
print(f"GFCCs (Placeholder - Not truly computed directly): {gfccs_mean_placeholder}")

spectral_centroids = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
spectral_centroid_mean = np.mean(spectral_centroids)
print(f"Spectral Centroid (mean): {spectral_centroid_mean:.2f} Hz")

times_spectral = librosa.times_like(spectral_centroids, sr=sr)
plt.figure(figsize=(12, 4))
plt.plot(times_spectral, spectral_centroids, color='purple', linewidth=2)
plt.title('Spectral Centroid Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.grid(True)
plt.tight_layout()
plt.show()

f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C5'), sr=sr)
avg_pitch_yin = np.nanmean(f0) if np.any(~np.isnan(f0)) else 0.0
print(f"Average Pitch (YIN): {avg_pitch_yin:.2f} Hz")

times_f0 = librosa.times_like(f0, sr=sr)
plt.figure(figsize=(12, 4))
plt.plot(times_f0[voiced_flag], f0[voiced_flag], '.', label='F0 (Voiced)', color='green')
plt.title('Pitch (F0) Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.grid(True)
plt.legend()
```
