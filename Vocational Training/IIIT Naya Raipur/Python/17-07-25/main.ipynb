# Load library

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
audio_file_path = "Lunch_Video_Generated_Successfully.mp3"
```

```python

try:
    y, sr = librosa.load(audio_file_path)
    print(f"Audio loaded successfully! Sampling Rate (sr): {sr} Hz")
    print(f"Audio duration: {len(y)/sr:.2f} seconds")

except FileNotFoundError:
    print(f"Error: Audio file not found at '{audio_file_path}'.")
    print("Please make sure the file is in the correct directory or provide the full path.")
    print("You can download a sample WAV file from: https://www.soundhelix.com/examples/mp3/SoundHelix-Song-1.mp3 (rename to .wav or convert)")
    exit() # Agar file nahi mili toh code ko rok dein
```

```python
plt.figure(figsize=(12, 4)) # Graph ka size set karein
librosa.display.waveshow(y=y, sr=sr, x_axis='time', color='blue')
plt.title('Audio Waveform') # Graph ka title
plt.xlabel('Time (s)') # X-axis ka label
plt.ylabel('Amplitude') # Y-axis ka label
plt.grid(True) # Grid lines dikhayein
plt.tight_layout() # Layout adjust karein
plt.show() # Graph display karein
```

```python
S = librosa.feature.melspectrogram(y=y, sr=sr)

# Power spectrogram ko decibel scale par convert karein (visulization ke liye better)
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(12, 4)) # Graph ka size set karein
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB') # Color bar add karein
plt.title('Mel Spectrogram') # Graph ka title
plt.xlabel('Time (s)') # X-axis ka label
plt.ylabel('Mel Frequency (Hz)') # Y-axis ka label
plt.tight_layout() # Layout adjust karein
plt.show() # Graph display karein
```
