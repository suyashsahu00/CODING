```python
# 1. Read the one-row CSV as a list and reshape into a DataFrame
import pandas as pd
import numpy as np

```

```python
# Read the file as a single row
with open('synthetic_students_one_row - synthetic_students_one_row.csv', 'r', encoding='utf-8') as f:
    data = f.read().strip().split(',')

```

```python
# Each student has 5 columns: Name, ID, Age, Gender, Address
num_cols = 5
num_students = len(data) // num_cols

```

```python
# Reshape into a 2D array
reshaped = np.array(data).reshape(num_students, num_cols)

```

```python
# Create DataFrame with column names
columns = ['Name', 'ID', 'Age', 'Gender', 'Address']
df = pd.DataFrame(reshaped, columns=columns)

```

```python
# Data cleaning: remove duplicates, handle nulls, reset index
import numpy as np

# Replace empty strings with NaN
df.replace('', np.nan, inplace=True)
# Remove duplicate rows
df.drop_duplicates(inplace=True)
# Remove rows with any NaN values
df.dropna(inplace=True)
# Reset the index of the DataFrame
df.reset_index(drop=True, inplace=True)
# Display the cleaned DataFrame
df.head()
```

```python
# Read the one-row CSV and reshape into a DataFrame
import pandas as pd
import numpy as np

with open('synthetic_students_one_row - synthetic_students_one_row.csv', 'r', encoding='utf-8') as f:
    data = f.read().strip().split(',')

num_cols = 5
num_students = len(data) // num_cols
reshaped = np.array(data).reshape(num_students, num_cols)
columns = ['Name', 'ID', 'Age', 'Gender', 'Address']
df = pd.DataFrame(reshaped, columns=columns)
df.head()
```

```python
# Data cleaning: remove duplicates, handle nulls, reset index
df.drop_duplicates(inplace=True)
df.replace('', np.nan, inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```
