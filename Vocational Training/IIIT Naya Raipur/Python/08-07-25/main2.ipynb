# Data Visualization 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt('subject_1.csv', delimiter=',', skip_header=1)

# Check shape and preview
print(data.shape)
print(data[:5])  # First 5 rows
```

```python
data = np.genfromtxt('subject_1.csv', delimiter=',', skip_header=1, usecols=(0, 1, 2))
```

```python

import matplotlib.pyplot as plt

x = data[:, 0]  # First column
y = data[:, 1]  # Second column

plt.plot(x, y, color='blue', linestyle='-', marker='o')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y, color='green', alpha=0.6)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.hist(data[:, 1], bins=20, color='orange', edgecolor='black')
plt.title('Histogram of Column 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(data[:, 1])
plt.title('Box Plot of Column 2')
plt.ylabel('Values')
plt.show()
```

```python
plt.imshow(data, cmap='viridis', aspect='auto')
plt.colorbar()
plt.title('Heatmap of CSV Data')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```
